```python
matches = {}
for index_small, row_small in all_imgs_df.iterrows():
    timestamp_small = row_small['Img_names']
    closest_match = None
    min_time_diff = float('inf')
    
    # Loop through the rows in the larger DataFrame
    for index_large, row_large in df_new.iterrows():
        timestamp_large = row_large['Name']
        time_diff = abs((timestamp_small - timestamp_large).total_seconds())
        
        # Update the closest match if this one is closer
        if time_diff < min_time_diff:
            min_time_diff = time_diff
            closest_match = timestamp_large
    
    # Store the closest match in the dictionary
    matches[timestamp_small] = closest_match

# Create a new DataFrame to store the matches
matches_df = pd.DataFrame(list(matches.items()), columns=['timestamp_small', 'closest_match'])
```

```python
# Scale for PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_pca)
```

```python
scaled_data = scaler.transform(df_pca)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
exp_var_pca = pca.explained_variance_ratio_
#
# Cumulative sum of eigenvalues; This will be used to create step plot
# for visualizing the variance explained by each principal component.
#
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
#
# Create the visualization plot
#
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
for i, evr in enumerate(exp_var_pca):
    print(f"Component {i+1}: {evr:.4f} ({evr*100:.2f}%)")
print(f"Cumulative explained variance (first {10} components): {cum_sum_eigenvalues[-1]:.4f} ({cum_sum_eigenvalues[-1]*100:.2f}%)")
```

#### Processing to extract the humidity and temperature values

```python
# Now, modelling the target variable 
# It has been recorded that at June 15 4:45PM drying was completed. 
dry_time = pd.to_datetime('2023-06-15 16:45:00')
```

```python
# Matching to extract the humidity and temp
hum_lis, temp_lis = [], []
for i in matches_df['closest_match']:
    hum, temp = df_new.loc[df_new['Name'] == i, ['Humidity', 'Temperature']].values[0]
    hum_lis.append(hum)
    temp_lis.append(temp)
print("Humidity list size: ", len(hum_lis))
print("Temperature list size: ", len(temp_lis))
```

```python
# Attaching the temp and hum values to the dataframe
pca_out_df['hum'] = hum_lis
pca_out_df['temp'] = temp_lis
pca_out_df
```

```python
# Calculate the y variable 
time_diff_lis = pca_out_df['Image_names'].apply(lambda x: max(0, (dry_time - pd.to_datetime(x)).total_seconds()))
time_diff_lis
```

```python
pca_out_df['target'] = time_diff_lis
pca_out_df = pca_out_df.drop('Image_names', axis=1)
pca_out_df
```

```python
# Saving the master regression dataframe
pca_out_df.to_csv('master_regression_dataset.csv', index = 'False')
```

```python
from numpy import absolute
from pandas import read_csv
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from xgboost import XGBRegressor
from sklearn.metrics import r2_score

# Load the dataset
data = pca_out_df.values

# Split data into input and output columns
X, y = data[:, :-1], data[:, -1]

# Define the XGBoost model
model = XGBRegressor()

# Define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# Evaluate model for MAE
scores_mae = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
scores_mae = absolute(scores_mae)

# Evaluate model for R-squared (R^2) score
scores_r2 = cross_val_score(model, X, y, scoring='r2', cv=cv, n_jobs=-1)

# Force MAE scores to be positive
scores_mae = absolute(scores_mae)

print('Mean MAE: %.3f (%.3f)' % (scores_mae.mean(), scores_mae.std()))
print('Mean R-squared (R^2) Score: %.3f (%.3f)' % (scores_r2.mean(), scores_r2.std()))
```
