```python
import numpy as np
import pandas as pd
import cv2
from PIL import Image
import os
import matplotlib.pyplot as plt
```

```python
image_directory_day1 = 'AutoML/TomatoDatasetPPM3/Day1 - 16.08.22/CLIPPED/'
images = os.listdir(image_directory_day1)
for i, image_name in enumerate(images):    #Remember enumerate method adds a counter and returns the enumerate object
    image_dataset_1 = []
    if (image_name.split('.')[1] == 'ppm'):
        file = open(image_directory_day1+image_name, "r")
        img_str = file.readlines()

        for j in range(3,len(img_str)):
            x, y, z = img_str[j].split(' ')
            image_dataset_1.append(int(x))
            image_dataset_1.append(int(y))
            image_dataset_1.append(int(z))
        
        image_dataset_1 = np.array(image_dataset_1)
        
        for k in range(0, len(image_dataset_1), 3):
            a = image_dataset_1[k]/image_dataset_1[k+2]
            b = image_dataset_1[k]/image_dataset_1[k+1]
            c = image_dataset_1[k+2]/image_dataset_1[k+1]
            
            if a > 0.75 and a < 1.25:
                if b > 0.75 and b < 1.25:
                    if c > 0.75 and c < 1.25:
                        image_dataset_1[k] = 0
                        image_dataset_1[k+1] = 0
                        image_dataset_1[k+2] = 0
            
        # Reconstructing the ppm file
        stri = "P3\n64 64\n255\n"
        for i in range(0, len(image_dataset_1), 3):
            temp = str(image_dataset_1[i])+' '+str(image_dataset_1[i+1])+' '+str(image_dataset_1[i+2])+'\n'
            stri+=temp
        
    fptr = open('AutoML/Tomato_background_subtracted/Day_1/'+str(image_name), 'w')
    fptr.write(stri)
    fptr.close()
```
